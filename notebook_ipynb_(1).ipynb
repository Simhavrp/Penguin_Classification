```python
import numpy as np

class LogisticRegression:
    def __init__(self, learning_rate=0.01, num_iterations=1000, weights=None):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = weights
        self.bias = 0

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z.astype('float')))

    def cost(self, w, b, x, y):
        m = x.shape[1]
        # number of examples

        # Compute the sigmoid function
        z = np.dot(x,w.T) + b
        h = 1 / (1 + np.exp(-z.astype('float')))

        # Compute the cost
        J = -1/m * np.sum(y * np.log(h) + (1-y) * np.log(1-h))
        return J


    def fit(self, X, y):
        # If weights are not provided, initialize them randomly
        if self.weights is None:
            self.weights = np.random.uniform(low=0, high=1, size=X.shape[1])

        # Gradient descent loop
        for i in range(self.num_iterations):
            z = np.dot(X, self.weights)
            y_hat = self.sigmoid(z.astype('float'))

            # Compute the gradient of the cost function
            gradient = np.dot(X.T, (y_hat - y)) / len(y)
#             print(len(self.weights.shape))
            Cost = self.cost(self.weights, self.bias, X, y)
            print(Cost)
#             Update weights
            self.weights = self.weights - self.learning_rate * gradient

    def predict(self, X):
        z = np.dot(X, self.weights)
        y_pred = self.sigmoid(z.astype('float'))
        y_label = np.where(y_pred>0.35,1,0)
        return y_pred,y_label
```
